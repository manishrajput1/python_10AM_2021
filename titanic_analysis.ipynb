```python
d=data.split('\n')
```

```python
for i in d[:-1]:
    j=i.split('\t\t')
    print(repr(j))
```

```python
death=0
survived=0
for i in d[:-1]:
    j=i.split('\t\t')
    if j[0]=='0':
        death+=1
    elif j[0]=='1':
        survived+=1
print('total death: ',death)
print('total survived: ',survived)
```

```python
male_survived=0
female_survived=0
male_death=0
female_death=0
for i in d[:-1]:
    j=i.split('\t\t')
    if j[0]=='1' and j[2]=='male':
        male_survived+=1
    elif j[0]=='1' and j[2]=='female':
        female_survived+=1
    elif j[0]=='0' and j[2]=='female':
        female_death+=1
    elif j[0]=='0' and j[2]=='male':
        male_death+=1
print('total male death: ',male_death)
print('total female death: ',female_death)
print()
print('total male survived: ',male_survived)
print('total female survived: ',female_survived)
```
